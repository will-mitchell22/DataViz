```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/drive/MyDrive/Top_Highest_Openings.csv'
df = pd.read_csv(file_path)

distributors = df['Distributor'].unique()
colors = ['green', 'blue', 'red', 'brown', 'yellow']

plt.figure(figsize=(12, 6))

for i, distributor in enumerate(distributors):
  distributor_df = df[df['Distributor'] == distributor]
  plt.scatter(
      distributor_df['Total Gross'],
      distributor_df['Distributor'],
      color=colors[i % len(colors)],
      alpha=0.5,
      label=distributor
  )

plt.xlabel('Total Gross')
plt.ylabel('Distributor')
plt.title('Total Gross by Distributor')
plt.legend(loc='upper right')
plt.show()
```
